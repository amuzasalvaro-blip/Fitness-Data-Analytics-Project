## 4.2 Tabla de hechos

La tabla de hechos es la tabla principal del modelo estrella y contiene la información de cada sesión de entrenamiento realizada. Además de las claves de relación con las dimensiones, almacena las métricas económicas utilizadas en el análisis del negocio.

Script SQL utilizado para la creación de la tabla de hechos fact_entrenamientos.
```sql

    sesion_id VARCHAR(15) PRIMARY KEY,
    fecha_id VARCHAR(15),

    margen_gimnasio NUMERIC(5,2)
```

Interpretación de los resultados

Los resultados muestran que el nivel físico alto es el más frecuente entre los clientes, con 755 usuarios, seguido del nivel medio con 596 y del nivel bajo con 247. Esto indica una mayor presencia de clientes con una condición física media o alta, mientras que los usuarios con un nivel físico bajo representan una proporción menor dentro de los gimnasios analizados.

## 6.8 Análisis predictivo mediante regresión lineal

### 6.8.1 Introducción y objetivo

La regresión lineal es una técnica estadística utilizada para analizar la relación existente entre una variable independiente y una variable dependiente. Su finalidad es identificar tendencias y estimar cómo los cambios producidos en una variable pueden influir sobre otra.

En este proyecto se utilizará una regresión lineal simple con el objetivo de estudiar la posible relación entre la experiencia de los entrenadores y el beneficio económico generado para el gimnasio. De esta forma, se pretende determinar si los entrenadores con un mayor número de años de experiencia tienden a generar mayores beneficios a través de las sesiones realizadas con los clientes.

Para ello se emplearán los datos almacenados en las tablas Fact_Entrenamientos y Dim_Entrenadores, utilizando como variable independiente los años de experiencia del entrenador y como variable dependiente el beneficio total generado.

### 6.8.2 Preparación de los datos

Para la construcción del modelo de regresión lineal fue necesario realizar una preparación previa de los datos. El análisis se centró en estudiar la relación entre la experiencia de los entrenadores y el beneficio económico generado para el gimnasio.

La variable independiente seleccionada fue el número de años de experiencia de cada entrenador, obtenido de la tabla Dim_Entrenadores. Por otro lado, la variable dependiente utilizada fue el beneficio total generado por cada entrenador, calculado a partir de los registros almacenados en la tabla Fact_Entrenamientos.

Dado que la tabla de hechos contiene información a nivel de sesión individual, fue necesario agrupar los datos por entrenador y calcular el beneficio total acumulado para cada uno de ellos. Posteriormente, esta información se combinó con los datos de experiencia de la dimensión de entrenadores, obteniendo un conjunto de datos adecuado para la aplicación del modelo de regresión lineal.

````python
import pandas as pd

archivo = "Sesiones de Entrenamientos Final.xlsx"

excel = pd.ExcelFile(archivo)

print(excel.sheet_names)
````

````python
fact = pd.read_excel(
    "Sesiones de Entrenamientos Final.xlsx",
    sheet_name="fact_entrenamientos"
)

entrenadores = pd.read_excel(
    "Sesiones de Entrenamientos Final.xlsx",
    sheet_name="dim_entrenadores"
)


print("FACT:")
print(fact.columns.tolist())

print("\nENTRENADORES:")
print(entrenadores.columns.tolist())
````

````python
beneficio_entrenador = fact.groupby("entrenador_id")["beneficio_gimnasio"].sum().reset_index()

df_regresion = beneficio_entrenador.merge(
    entrenadores[["entrenador_id", "anios_experiencia"]],
    on="entrenador_id"
)

df_regresion.head()
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(
    df_regresion["anios_experiencia"],
    df_regresion["beneficio_gimnasio"]
)

plt.xlabel("Años de experiencia")
plt.ylabel("Beneficio total generado")
plt.title("Relación entre experiencia y beneficio")

plt.show()
````

El gráfico de dispersión muestra una tendencia positiva entre los años de experiencia de los entrenadores y el beneficio total generado. Aunque existe cierta dispersabilidad en los datos, se observa que los entrenadores con mayor experiencia tienden a generar beneficios superiores, lo que justifica la aplicación de un modelo de regresión lineal para estudiar esta relación de forma más precisa.

### 6.8.3 Construcción del modelo

Una vez preparados los datos, se procede a la construcción del modelo de regresión lineal simple utilizando la librería Scikit-Learn de Python. Se definió como variable independiente (X) los años de experiencia de los entrenadores y como variable dependiente (Y) el beneficio total generado. Posteriormente, el modelo fue entrenado con el fin de estimar la relación existente entre ambas variables y obtener los parámetros de la recta de regresión.

````python
from sklearn.linear_model import LinearRegression

# Variable independiente
X = df_regresion[["anios_experiencia"]]

# Variable dependiente
y = df_regresion["beneficio_gimnasio"]

# Crear modelo
modelo = LinearRegression()

# Entrenar modelo
modelo.fit(X, y)

print("Pendiente:", modelo.coef_[0])
print("Intercepto:", modelo.intercept_)
print("R²:", modelo.score(X, y))
````

### 6.8.4 Resultados obtenidos

````python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

# Estilo
sns.set_theme(style="whitegrid")

# Crear figura
fig, ax = plt.subplots(figsize=(14, 8))

# Regresión
sns.regplot(
    data=df_regresion,
    x="anios_experiencia",
    y="beneficio_gimnasio",
    scatter_kws={
        "s": 70,
        "alpha": 0.85,
        "edgecolor": "#1d4ed8",
        "linewidths": 1
    },
    line_kws={
        "color": "crimson",
        "linewidth": 3
    },
    color="crimson",
    ax=ax
)

# Título principal
ax.set_title(
    "Relación entre la experiencia de los entrenadores y el beneficio generado",
    fontsize=24,
    fontweight="bold",
    color="#0b1b4d",
    pad=40
)

# Subtítulo
ax.text(
    0.5,
    1.03,
    "Regresión lineal simple",
    transform=ax.transAxes,
    ha="center",
    fontsize=18,
    style="italic",
    color="#5c6b8a"
)

# Ejes
ax.set_xlabel(
    "Años de experiencia",
    fontsize=18,
    fontweight="bold",
    color="#0b1b4d"
)

ax.set_ylabel(
    "Beneficio total generado (€)",
    fontsize=18,
    fontweight="bold",
    color="#0b1b4d"
)

# Caja estadísticas
texto = (
    f"Pendiente: {modelo.coef_[0]:.2f}\n\n"
    f"Intercepto: {modelo.intercept_:.2f}\n\n"
    f"R²: {modelo.score(X,y):.3f} ({modelo.score(X,y)*100:.1f}%)"
)

ax.text(
    0.05,
    0.78,
    texto,
    transform=ax.transAxes,
    fontsize=14,
    bbox=dict(
        boxstyle="round,pad=0.8",
        facecolor="white",
        edgecolor="#cbd5e1",
        alpha=0.95
    )
)

# Leyenda personalizada
legend_elements = [
    Line2D(
        [0], [0],
        marker='o',
        color='w',
        markerfacecolor='#3b82f6',
        markeredgecolor='#1d4ed8',
        markersize=10,
        label='Datos reales'
    ),
    Line2D(
        [0], [0],
        color='crimson',
        lw=3,
        label='Recta de regresión'
    ),
    Patch(
        facecolor='lightcoral',
        alpha=0.25,
        label='Intervalo de confianza (95%)'
    )
]

ax.legend(
    handles=legend_elements,
    loc="lower right",
    fontsize=13,
    frameon=True
)

# Cuadrícula
ax.grid(True, linestyle="--", alpha=0.4)

# Limpiar bordes superiores
sns.despine()

plt.tight_layout()
plt.show()
````

La aplicación del modelo de regresión lineal permite identificar una relación positiva entre los años de experiencia de los entrenadores y el beneficio generado para el gimnasio. ## 6.9 Segmentación de entrenadores mediante K-Means

### 6.9.1 Introducción y objetivo

En este proyecto se utiliza K-Means para segmentar a los entrenadores según su experiencia, valoración y beneficio generado. De esta forma, se busca identificar diferentes perfiles de entrenadores que puedan resultar útiles para el análisis y la toma de decisiones dentro del gimnasio.

### 6.9.2 Preparación de los datos

Para aplicar el algoritmo K-Means fue necesario crear un conjunto de datos específico para los entrenadores. Para ello, se combinaron los datos de beneficio total generado con la información de experiencia y valoración media de cada entrenador.

Las variables seleccionadas fueron los años de experiencia, la valoración media recibida por los clientes y el beneficio generado para el gimnasio. Estas variables permiten analizar tanto el rendimiento económico como las características profesionales de los entrenadores.


````python
df_kmeans_entrenadores = beneficio_entrenador.merge(
    entrenadores[
        ["entrenador_id", "anios_experiencia", "rating_medio"]
    ],
    on="entrenador_id"
)

df_kmeans_entrenadores.head()
````

Selección de variables que serán utilizadas por el algoritmo. Años de experiencia, valoracion de los entrenadores y el beneficio generado para los gimnasios.

La selección de estas variables permite que la segmentación tenga en cuenta tanto aspectos relacionados con la experiencia profesional como indicadores de rendimiento y satisfacción de los clientes.

````python
variables = df_kmeans_entrenadores[
    ["anios_experiencia", "rating_medio", "beneficio_gimnasio"]
].copy()
````

### 6.9.3 Aplicación del algoritmo K-Means

Antes de aplicar el algoritmo K-Means fue necesario realizar una normalización de los datos mediante la técnica StandardScaler. Este proceso transforma las variables a una misma escala, evitando que aquellas con valores más elevados tengan una influencia mayor en la formación de los grupos.

Se verificó el formato numérico de la variable de valoración media para garantizar su correcto tratamiento durante el análisis.


````python
from sklearn.preprocessing import StandardScaler

# Convert 'rating_medio' to numeric, handling comma decimal separator
variables['rating_medio'] = variables['rating_medio'].astype(str).str.replace(',', '.', regex=False).astype(float)

scaler = StandardScaler()

variables_escaladas = scaler.fit_transform(variables)
````

Una vez normalizados los datos, se utilizó el método del codo para determinar el número óptimo de grupos.

Como puede observarse en la figura, la reducción de la inercia es significativa hasta alcanzar tres grupos. A partir de este punto, la mejora obtenida al aumentar el número de clusters es menor, por lo que se seleccionó un valor de k = 3 para realizar la segmentación de los entrenadores.


````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inercias = []

for k in range(1, 11):
    modelo = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    modelo.fit(variables_escaladas)
    inercias.append(modelo.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), inercias, marker="o")

plt.title("Método del codo - Entrenadores")
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")

plt.grid(True)

plt.show()
````

Una vez determinado el número óptimo de grupos mediante el método del codo, se aplicó el algoritmo K-Means utilizando tres clusters. El modelo asignó automáticamente cada entrenador al grupo con características más similares, teniendo en cuenta los años de experiencia, la valoración media y el beneficio generado.

Como resultado, cada entrenador quedó clasificado dentro de uno de los tres perfiles identificados por el algoritmo, lo que permitió realizar posteriormente un análisis comparativo entre los distintos grupos obtenidos.

````python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

df_kmeans_entrenadores["cluster"] = kmeans.fit_predict(
    variables_escaladas
)

df_kmeans_entrenadores.head()
````

### 6.9.4 Resultados obtenidos

Se obtiene tres grupos diferenciados de entrenadores.

Los resultados muestran que el Cluster 0 presenta los valores medios más elevados tanto en experiencia como en beneficio generado, lo que indica que está formado por los entrenadores con mejor rendimiento. Por otro lado, el Cluster 1 agrupa a los entrenadores con menor experiencia y menor beneficio medio. Finalmente, el Cluster 2 presenta valores intermedios, representando perfiles en proceso de consolidación profesional.

````python
df_kmeans_entrenadores.groupby("cluster").mean(
    numeric_only=True
)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df_kmeans_entrenadores,
    x="anios_experiencia",
    y="beneficio_gimnasio",
    hue="cluster",
    s=120
)

plt.title(
    "Segmentación de entrenadores mediante K-Means",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Años de experiencia")
plt.ylabel("Beneficio total generado (€)")

plt.legend(title="Cluster")

plt.grid(alpha=0.3)

plt.show()
````

Los resultados obtenidos muestran que los entrenadores pueden clasificarse en tres grupos principales según sus características y rendimiento. El primer grupo está formado por entrenadores con una mayor experiencia y una generación de beneficios superior a la media. El segundo grupo reúne perfiles con menor experiencia y menor contribución económica. Finalmente, el tercer grupo representa entrenadores con resultados intermedios.

La segmentación obtenida puede servir como apoyo para la toma de decisiones relacionadas con la asignación de clientes, la planificación de programas de formación o el diseño de estrategias orientadas a mejorar el rendimiento de los entrenadores con mayor potencial de crecimiento.
